```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import r2_score, mean_absolute_error
import numpy as np
```

```python
df = pd.read_csv(r'yahoo_stock.csv')
df
```

```python
dados = df['High'].values
tam = len(dados)

tam_teste = int(tam * 0.7)
tam_train = tam - tam_teste

valores_treino = dados[tam_teste:]
valores_teste = dados[:tam_train]
```

```python
janela = 10

X_train = []
y_train = []

for i in range(len(valores_treino) - janela - 1):
    X_train.append(valores_treino[i:i + janela])
    y_train.append(valores_treino[i + janela])

X_train = np.array(X_train)
y_train = np.array(y_train)

X_teste = []
y_teste = []

for i in range(len(valores_teste) - janela - 1):
    X_teste.append(valores_teste[i:i + janela])
    y_teste.append(valores_teste[i + janela])

X_teste = np.array(X_teste)
y_teste = np.array(y_teste)
```

Umas das principais diferenças entre as formas apresentadas de ML, é como os dados chegam a rede, no final o objetivo é prever valores de uma mesma coluna porém as formas são completamente diferentes, a forma desta rede é baseada no metodo de janelas, onde se cria um intervalo de tempo e se tenta prever o proximo valor da sequencia. Já a outra forma abordada se baseia em pegar todas as informações do DF e tentar prever os valores da coluna target.

```python
model = Sequential()
model.add(Dense(30, activation='relu'))
model.add(Dense(1, activation='linear'))
model.compile(optimizer='adam', loss='huber')
```

```python
model.fit(X_train, y_train, epochs=20, batch_size=10, verbose=1)
```

```python
pred = model.predict(X_teste)
pred
```

```python
r2 = r2_score(y_teste, pred)
print("R-quadrado:", r2)

mae = mean_absolute_error(y_teste, pred)
print("MAE:", mae)
```
